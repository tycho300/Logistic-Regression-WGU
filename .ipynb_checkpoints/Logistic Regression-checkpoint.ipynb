### Research Question

What factors contribute to the probability that a patient will be readmitted to the hospital?


### Goals of Analysis

#### Organizational Goals 
The data for the medical dataset provided was created in response for the need to understand the causes of patient readmission. Hospitals are fined for high rates of readmission and identifying potential causes for readmission can help hospitals mitigate those fines. For example, if a particular health condition is associated with high rates of readmission those patients with that condition could receive additional assistance at the time of discharge for outpatient monitoring which could mitigate the potential for readmission. The goal of this analysis is to search for any factors that are associated with and can predict the likelihood of readmission so that the knowledge can be used to identify ways of mitigating the causes of readmission. The hospital owners and administrators would directly benefit from this analysis in the form of reduced fines from high readmission rates as well as reduced costs associated with repeat hospital visits. Secondly patients can benefit from reducing the financial and physical stresses of repeated hospital stays. 

#### Logistical Goals

The overall goal of this analysis is to perform logistic regression models which identifies factors that can predict if a patient will be readmitted to the hospital. The dataset provided will be cleaned and all categorical variables will be encoded in ways that contribute to the construction of the model. All variables under consideration will be examined with descriptive statistics, univariate and bivariate plots which examine their relationship with the target variable for readmission. An initial logistical regression model will be constructed with all variables. From there, using statistical methods, the primary independent variables which best explain the probability of readmission will be identified. A final reduced model will be constructed. The two models will be compared and discussed. 

### Assumptions of Logistic Regression Model

**<u>First Assumption</u>** The target variable is a categorical variable with binary, 0 and 1, values. 

**<u>Second Assumption</u>** There is low or no multicollinearity between the independent variables. There are no independent variables which are highly correlated with one another.

**<u>Third Assumption</u>** There is a linear relationship between the log-odd function of the target variable and the dependent variable. (Statology, 2020)

**<u>Fourth Assumption</u>** The sample size, or number of observations is large.



### Data cleaning

Any missing values will be imputed with statistical methods that maintain the overall distribution of the original dataset and supports the accuracy of the analysis. Duplicate values will be removed so that there isn't a compounding effect of similar values in the results. Outliers will be treated in a way that supports the natural variance in the data and does not force statistical relationships that do not exist. Anomalous values will either be dropped or replaced with appropriate values if they can be estimated.

### Initial logistical regression model

```python
# Print column names for easy cut and paste 
lrmodeldf.columns
```

```python
# Run initial logistic regression model using statsmodels Logit function
import statsmodels.api as stats
independent = lrmodeldf[['Population', 'Age', 'Income', 'VitD_levels', 'Initial_days',
       'Full_meals_eaten', 'HighBlood', 'Complication_risk_High',
       'Complication_risk_Low', 'Services_Blood Work', 'Services_CT Scan',
       'Services_MRI']].astype(float)
independent = stats.add_constant(independent)

dependent = lrmodeldf['ReAdmis']

model_initial=stats.Logit(dependent,independent).fit()

model_initial.summary()
```

### Feature Selection 

The feature selection method chosen was an iterative process which removes features step-wise until a reduced model is found. The main justification for this process is the use of statistical measurements to ensure accuracy of the final model by eliminating features that are less useful statistically. 

The first step will be to address issues of multicollinearity. It is an assumption of logistic regression models that any independent variables will not be highly correlated. Therefore, any independent variables found to have a high degree of multicollinearity will be removed from model. 

The next step in this process will be to remove any statistically insignificant features with p-values greater than .05. 

The final step will be an iterative process to remove each feature piecewise to determine which combination of variables maintain the accuracy of the model but reduce the features to the simplest equation. 

### Reducing the model by checking for multicollinearity 
Using the Variance Inflation Factor, independent variables will be identified that have a high degree of multicollinearity. Variance Inflation Factor is a measure of how much variance is inflated by correlation between independent variables. 

```python
# VIF of Independent variables to check for multicollinearity (Geeks for Geeks, n.d.)
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = lrmodeldf[['Population', 'Age', 'Income', 'VitD_levels', 'Initial_days',
       'Full_meals_eaten', 'HighBlood', 'Complication_risk_High',
       'Complication_risk_Low', 'Services_Blood Work', 'Services_CT Scan',
       'Services_MRI']].astype(float)

# Run VIF for independent variables
{X.columns[i]: variance_inflation_factor(X.values, i) for i in range(0, X.shape[1])}
```

The only value with a problematically high VIF value is 'VitD_levels' at 12.73. While cutoff values for VIF are debated it is generally accepted that anything over 10 is problematic and indicates a high degree of collinearity. The model will be run without this feature. (Choueiry, n.d.)

```python
# Run model without features with high values for VIF 

independent = lrmodeldf[['Population', 'Age', 'Income', 'Initial_days',
       'Full_meals_eaten', 'HighBlood', 'Complication_risk_High',
       'Complication_risk_Low', 'Services_Blood Work', 'Services_CT Scan',
       'Services_MRI']].astype(float)

independent = stats.add_constant(independent)

dependent = lrmodeldf['ReAdmis']

model_step1=stats.Logit(dependent,independent).fit()

model_step1.summary()
```

### Reducing model by checking for statistical significance
The variables, 'Population', 'Age', 'Income', 'Full_meals_eaten', and 'Services_Blood Work'  all have p-values much greater than .05. They will be removed from the model at this next step. 

```python
# Run model after removing features with high p-values

independent = lrmodeldf[['Initial_days','HighBlood', 'Complication_risk_Low', 
                         'Services_CT Scan','Services_MRI']].astype(float)

independent = stats.add_constant(independent)

dependent = lrmodeldf['ReAdmis']

model_step2=stats.Logit(dependent,independent).fit()

model_step2.summary()
```

### Iterative model reduction by calculating model accuracy

The final step will be to remove features in stages, comparing the accuracy of the model at each stage to determine which reduced model is the most accurate. The calculation for accuracy will be determined from the confusion matrix and dividing the total correct predictions divided by all predictions. 

```python
# Calculate accuracy of the last stage of model reduction to compare at next phase 
model = model_step2
cm = model.pred_table()
row1 = cm[0]
row2 = cm[1]
tn_2 = row1[0] #True Negative
fp_2 = row1[1] #False Positive
fn_2 = row2[0] #False Negative
tp_2 = row2[1] #True Positive
accuracy_2 = (tn_2 + tp_2) / (tn_2 + fp_2 + fn_2 + tp_2)
print("The accuracy of the model is ", accuracy_2)
```

The model is currently 98 percent accurate at predicting readmission outcomes. The next step will be to remove a feature to see if the accuracy improves or declines. 

```python
# Remove feature 'Services_MRI' and run model & calculate accuracy

independent = lrmodeldf[['Initial_days','HighBlood', 'Complication_risk_Low', 
                         'Services_CT Scan']].astype(float)
independent = stats.add_constant(independent)

dependent = lrmodeldf['ReAdmis']

model_step3=stats.Logit(dependent,independent).fit()

model = model_step3
cm = model.pred_table()
row1 = cm[0]
row2 = cm[1]
tn_3 = row1[0] #True Negative
fp_3 = row1[1] #False Positive
fn_3 = row2[0] #False Negative
tp_3 = row2[1] #True Positive
accuracy_3 = (tn_3 + tp_3) / (tn_3 + fp_3 + fn_3 + tp_3)
print("The accuracy of the model is ", accuracy_3)
print("The accuracy has changed ", accuracy_3 - accuracy_2)
```

The accuracy of the model has been reduced by .0003 which does not significantly reduce the model's efficacy. The feature 'Services_MRI' will not be added back into the model and 'Services_CT Scan' will be removed. 

```python
# Remove feature 'Services_CT' and run model & calculate accuracy

independent = lrmodeldf[['Initial_days','HighBlood', 'Complication_risk_Low']].astype(float)
independent = stats.add_constant(independent)

dependent = lrmodeldf['ReAdmis']

model_step4=stats.Logit(dependent,independent).fit()

model = model_step4
cm = model.pred_table()
row1 = cm[0]
row2 = cm[1]
tn_4 = row1[0] #True Negative
fp_4 = row1[1] #False Positive
fn_4 = row2[0] #False Negative
tp_4 = row2[1] #True Positive
accuracy_4 = (tn_4 + tp_4) / (tn_4 + fp_4 + fn_4 + tp_4)
print("The accuracy of the model is ", accuracy_4)
print("The accuracy has changed ", accuracy_4 - accuracy_3)
```

The accuracy of the model has been reduced by .001 which does not significantly reduce the model's efficacy. The feature 'Services_CT Scan' will not be added back into the model and 'Complication_risk_Low' will be removed.

### Confusion Matrix Comparison of Models

The best way to compare the efficacy of the models is by constructing a confusion matrix of predicted values against actual values. The values in this matrix can also be used to calculate other measures of comparison like accuracy, sensitivity, and specificity. 


```python
# Calculate confusion matrix for both initial and final models

cm_initial = model_initial.pred_table()
cm_final = model_final.pred_table()

print("The confusion matrix for the initial model is:\n", cm_initial)
print("The confusion matrix for the final model is:\n ", cm_final)
```

### Explanation of confusion matrix 

The values in the confusion matrix represent the true negative and true positive values on the diagonal starting on the upper left corner and the false values on the cross-wise diagonal. 

Upper Left = True Negatives
Upper Right = False Positives
Lower Left = False Negatives
Lower Right = True Postives 

The initial model has a higher number of accurate predictions compared to the final model. However, the overall proportion of accurate predictions to inaccurate predictions remains relatively small for both models. This can be seen in the following mosaic plots of the initial and final models. 

```python
# Display mosaic plots demonstrating the overall accuracy of Initial and Final models
from statsmodels.graphics.mosaicplot import mosaic

plt.rcParams["figure.figsize"] = [6, 6]

mosaic(cm_initial, title = "Initial Model Accuracy")
mosaic(cm_final, title = "Final Model Accuracy")
```

### Other Comparison Metrics 

The values in the confusion matrix can be used to calculate a number of metrics which help gauge the efficacy of the models. The first and most important is accuracy which is the proportion of correct predictions divided by the total number of predictions. This metric was used earlier in the model reduction phase as a means of measuring which models resulted in more accurate predictions


```python
# Calculate accuracy of initial model
cm_initial = model_initial.pred_table()
row1 = cm_initial[0]
row2 = cm_initial[1]
tn_i = row1[0] #True Negative
fp_i = row1[1] #False Positive
fn_i = row2[0] #False Negative
tp_i = row2[1] #True Positive
accuracy_initial = ((tn_i + tp_i) / (tn_i + fp_i + fn_i + tp_i))*100
print("The accuracy of the initial model is  ", accuracy_initial, " percent.")

# Calculate accuracy of the final model
cm_final = model_final.pred_table()
row1 = cm_final[0]
row2 = cm_final[1]
tn_f = row1[0] #True Negative
fp_f = row1[1] #False Positive
fn_f = row2[0] #False Negative
tp_f = row2[1] #True Positive
accuracy_final = ((tn_f + tp_f) / (tn_f + fp_f + fn_f + tp_f))*100
print("The accuracy of the final model is  ", accuracy_final, " percent.")
```

### Comparison of Accuracy

Both models show a high degree of accuracy, predicting the outcome of the target variable, at a 97% success rate. The reduction of the model from 12 to 2 variables did not significantly alter the efficacy of the prediction. It was an acceptable trade off for simplicity.

While the specificity or rate of true negatives did not change much from the inital to the final model, the specificity saw a small reduction. This means that the final model shows a very small reduction in the ability to predict true positives.  

### Summary of analysis

### Regression equation 
The regression equation is the form 
$$
\log(\frac{p}{1-p}) = a = bX
$$


Where b is the coefficient for Initial Days and X is the value of Initial Days and a is the constant or y intercept. 


$$
  \log(\frac{p}{1-p})=-53.59 + .9865(Initial Days)
$$ 
(Sefidian, n.d.)

### Interpretation of the coefficients

The coefficient b associated with the predictor variable 'Initial_days' is the change in log-odds for a one day increase in 'Initial_days'. It can also be intepreted by the odds ratio where each day increase in 'Initial_days' multiplies the odds-ratio of readmission by e^b or 2.46. The b coefficient is essentially the slope or rate of change of the log-odds equation as X changes. And the constant 'a' is the y-intercept or the log odds of a patient with zero days. (sefidian, n.d.)

### Statistical and practical significance of reduced model 

The statistical significance of the relationship between Initial_days and Readmission is strong. The final model does not include any independent variables which have a high degree of correlation which meets the statstical assumption of logistic regression. The p-values of the independent variable is 0 which indicates statistically significance. The pseudo-R-square value for the final model also shows a high explanation for the variance. Additionally, the regression equation has a high success of predicting the correct outcomes. 

Practically speaking the model is useful in that it tells stakeholders that there is a statistically significant relationship between Intial_days and Readmission. Knowing a patient is likely to be readmitted after a lengthy hospital stay identifies areas for improvement in care such as outpatient follow-up and home care which could monitor the patient's progress and help prevent readmission. 


### Limitations of data analysis 

The primary limitation of this analysis is the problem of correlation and causation. There appears to be a clear relationship between long initial stays and high probabilities of readmission. Unfortunately, this doesn't really tell us much. Kicking people out of the hospital early, before they are stable isn't likely to result in lower probabilities of readmission. The causitive relationship is not even suggested with this analysis.
